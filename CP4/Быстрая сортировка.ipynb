```python
import random
import time
```

```python
def partition(arr, low, high):
 
    pivot = arr[low]
    i = low - 1
    j = high + 1
 
    while True:
        i += 1
        while (arr[i] < pivot):
            i += 1
        j -= 1
        while (arr[j] > pivot):
            j -= 1
        if (i >= j):
            return j
 
        arr[i], arr[j] = arr[j], arr[i]

def quickSort(arr, low, high):
    if high - low < 1:
        pi = partition(arr, low, high)
        quickSort(arr, low, pi)
        quickSort(arr, pi + 1, high)
    return arr
```

```python
# Задаем константные размеры массивов

SMALL_ARR_SIZE = 20
MEDIUM_ARR_SIZE = 1000
BIG_ARR_SIZE = 10000
```

```python
# Массивы лучшего качества = best quality arrays

bq_small_arr = list(range(1, SMALL_ARR_SIZE + 1)) # +1, т.к. range() не включает значение конечного диапозона
bq_medium_arr = list(range(1, MEDIUM_ARR_SIZE + 1))
bq_big_arr = list(range(1, BIG_ARR_SIZE + 1))
```

```python
# Изучаем на массивах лучшего качества

# Исследуем малый массив высшего качества
print(bq_small_arr)

tic = time.perf_counter() # Засекаем время

bq_small_arr = quickSort(bq_small_arr, 0, SMALL_ARR_SIZE - 1)

print("Результат сортировки: ")
print(bq_small_arr)

toc = time.perf_counter()
print(f"Сортировка заняла {toc-tic:0.4f} секунд")
```

```python
# Исследуем средний массив высшего качества
print(bq_medium_arr)

tic = time.perf_counter() # Засекаем время

bq_medium_arr = quickSort(bq_medium_arr, 0, MEDIUM_ARR_SIZE - 1)

print("Результат сортировки: ")
print(bq_medium_arr)

toc = time.perf_counter()
print(f"Сортировка заняла {toc-tic:0.4f} секунд")
```

```python
# Исследуем большой массив высшего качества
print(bq_big_arr)

tic = time.perf_counter() # Засекаем время

bq_big_arr = quickSort(bq_big_arr, 0, BIG_ARR_SIZE - 1)

print("Результат сортировки: ")
print(bq_big_arr)

toc = time.perf_counter()
print(f"Сортировка заняла {toc-tic:0.4f} секунд")
```

```python
# Массивы среднего качества = average quality arrays

# Количество элементов для перемешиваня в среднем качестве
elements_to_shuffle = 0
# Процент элементов массива, что должен быть перемешан
PERCENT_ARR_TO_SHUFFLE = 0.1 #10%
# Индексы элементов массива, которые будут меняться местами
index1 = 0
index2 = 0


aq_small_arr = list(range(1, SMALL_ARR_SIZE + 1))
elements_to_shuffle = int(SMALL_ARR_SIZE * PERCENT_ARR_TO_SHUFFLE)

for i in range(elements_to_shuffle):
  # -1, т.к. нумерация массивов идёт с нуля
  index1 = random.randint(0, SMALL_ARR_SIZE - 1)
  index2 = random.randint(0, SMALL_ARR_SIZE - 1)

  aq_small_arr[index1], aq_small_arr[index2] = aq_small_arr[index2], aq_small_arr[index1]


aq_medium_arr = list(range(1, MEDIUM_ARR_SIZE + 1))
elements_to_shuffle = int(MEDIUM_ARR_SIZE * PERCENT_ARR_TO_SHUFFLE)

for i in range(elements_to_shuffle):
  index1 = random.randint(0, MEDIUM_ARR_SIZE - 1)
  index2 = random.randint(0, MEDIUM_ARR_SIZE - 1)

  aq_medium_arr[index1], aq_medium_arr[index2] = aq_medium_arr[index2], aq_medium_arr[index1]

aq_big_arr = list(range(1, BIG_ARR_SIZE + 1))
elements_to_shuffle = int(BIG_ARR_SIZE * PERCENT_ARR_TO_SHUFFLE)

for i in range(elements_to_shuffle):
  index1 = random.randint(0, BIG_ARR_SIZE - 1)
  index2 = random.randint(0, BIG_ARR_SIZE - 1)

  aq_big_arr[index1], aq_big_arr[index2] = aq_big_arr[index2], aq_big_arr[index1]

  
print(aq_small_arr)
print(aq_medium_arr)
print(aq_big_arr)
```
